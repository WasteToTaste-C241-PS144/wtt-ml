# Inisialisasi Library



```
Calls the required libraries, then initializes the dataset.
```



````python
import pandas as pd
````

````python
data_ayam = pd.read_csv('/content/dataset-ayam.csv').head(20)
data_ikan = pd.read_csv('/content/dataset-ikan.csv').head(20)
data_kambing = pd.read_csv('/content/dataset-kambing.csv').head(20)
data_sapi = pd.read_csv('/content/dataset-sapi.csv').head(20)
data_tahu = pd.read_csv('/content/dataset-tahu.csv').head(20)
data_telur = pd.read_csv('/content/dataset-telur.csv').head(20)
data_tempe = pd.read_csv('/content/dataset-tempe.csv').head(20)
data_udang = pd.read_csv('/content/dataset-udang.csv').head(20)

all_data = data_ayam._append([data_ikan, data_kambing, data_sapi, data_tahu, data_telur, data_tempe, data_udang])
all_data = all_data.reset_index(drop=True)
all_data.head()
````

````python
all_data.info()
````

# Generate Base Data

## Pengecekan Missing Values



```
Checking and handling missing values
```



````python
data_miss = all_data.isnull().sum()
print(data_miss)
````

````python
all_data.dropna(inplace=True)
all_data.reset_index(drop=True, inplace=True)
all_data.isnull().sum()
````

````python
all_data.info()
````

## Export Base Data



```
Exporting Database
```



````python
all_data.to_csv('recipes.csv', sep=',', index=False, header=True, encoding='utf-8')
````

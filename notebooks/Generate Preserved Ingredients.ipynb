# Initialization Library

```python
import pandas as pd
```

```python
# Load the recipes dataset from the provided URL
recipes = pd.read_csv('https://raw.githubusercontent.com/pijarpahlawan/WasteToTaste/master/complete_recipe_data.csv')
recipes.head()
```

```python
# Download the clean ingredients list from the provided URL
!wget https://raw.githubusercontent.com/pijarpahlawan/WasteToTaste/master/clean_ingredients.txt
```

```python
# Set the path to the downloaded clean ingredients file
clean_ingredients_url = '/content/clean_ingredients.txt'

# Read the content of the clean ingredients file
with open(clean_ingredients_url, 'r') as file:
    content = file.read()
```

```python
# Split the content into a list of cleaned words
cleaned_words = content.split()
print(cleaned_words)
```

```python
# Function to clean the ingredients in each recipe
def clean_ingredients(ingredients):
    # Split ingredients by '--'
    ingredients = ingredients.split('--')
    # Split each ingredient into words
    ingredient_words = [ingredient.split() for ingredient in ingredients]
    # Keep only the words that are in the cleaned words list
    cleaned_ingredients = [[word for word in words if word in cleaned_words] for words in ingredient_words]
    # Join the cleaned words back into a single string
    cleaned_ingredients = [' '.join(words) for words in cleaned_ingredients if words]
    return cleaned_ingredients
```

```python
# Apply the clean_ingredients function to the 'Ingredients' column
ingredients = recipes['Ingredients'].apply(clean_ingredients)
ingredients.head()
```

```python
# Initialize an empty list to store all cleaned ingredients
clean_ingredients_list = []
data = pd.DataFrame(ingredients)

# Collect all cleaned ingredients into a single list
for index, row in data.iterrows():
    clean_ingredients_list = clean_ingredients_list + row['Ingredients']

clean_ingredients_list
```

```python
# Remove duplicates by converting the list to a dictionary and back to a list
clean_ingredients_list = list(dict.fromkeys(clean_ingredients_list))
clean_ingredients_list
```

```python
# Convert the cleaned ingredients list to JSON format
json_format = [{"name": ingredient} for ingredient in clean_ingredients_list]
len(json_format)
```

```python
import json

# Save the cleaned ingredients list in JSON format to a file
with open('preserved_ingredients.json', 'w') as json_file:
    json.dump(json_format, json_file, indent=4)
```
